```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 50)
y = x;
plt.title("y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.plot(x,y);
```

```python
x = np.linspace(0, 10, 50)
y = x;
plt.title("y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(x,y, "r--");
```

```python
x = np.linspace(0, 10, 50)
y1 = x;
y2 = [i**2 for i in x]
plt.title("y1 = x, y2 = x^2")
plt.xlabel("x")
plt.ylabel("y1, y2")
plt.grid()

plt.plot(x,y1, label='$y1=x$');
plt.plot(x,y2, label='$y2=x^2$');
plt.legend()
```

```python
x = np.linspace(0, 10, 50)
y1 = x;
y2 = [i**2 for i in x]

plt.figure(figsize=(9,9))
plt.subplot(2, 1, 1)
plt.plot(x, y1, label='$y2=x^2$')
plt.title("y1 = x, y2 = x^2")
plt.ylabel("y1", fontsize = 14)
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, y2, label='$y2=x^2$')
plt.ylabel("y2", fontsize = 14)
plt.xlabel("x", fontsize = 14)
plt.grid(True)
plt.legend()
```
